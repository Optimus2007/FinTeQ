## 1) Setup

```python
from pathlib import Path
import json
import sys

ROOT = Path('..').resolve()
sys.path.append(str(ROOT / 'src'))

from level2_secondary.qml_extension import run_level2_qml_extension

TEMPLATE_PATH = ROOT / 'inputs' / 'test_template.xlsx'
OUT_XLSX = ROOT / 'results' / 'test_template_filled_hybrid_q80.xlsx'
OUT_REPORT = ROOT / 'results' / 'level2_hard_cv_q80_report.json'

print('Template:', TEMPLATE_PATH)
print('Template exists:', TEMPLATE_PATH.exists())
print('Output XLSX:', OUT_XLSX)
print('Output report:', OUT_REPORT)
```

## 2) Run Level-2

```python
# Same simple pattern as Level-1 notebook: set key params and run.
MIN_QUANTUM_WEIGHT = 0.80

report = run_level2_qml_extension(
    template_path=str(TEMPLATE_PATH),
    output_xlsx=str(OUT_XLSX),
    output_report=str(OUT_REPORT),
    min_quantum_weight=MIN_QUANTUM_WEIGHT,
    cv_repeats=3,
)

report['hybrid_params']
```
