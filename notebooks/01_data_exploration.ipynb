# 01 Data Exploration (Level-1 + Level-2)
This notebook performs a quick statistical check for both challenge datasets.

**Dataset source (both Level-1 and Level-2):** Hugging Face Hub — https://huggingface.co/datasets/Quandela/Challenge_Swaptions

```python
from datasets import load_dataset
import pandas as pd
import numpy as np

# Load both challenge datasets
level1_df = load_dataset(
    "Quandela/Challenge_Swaptions",
    data_files="level-1_Future_prediction/train.csv",
    split="train",
).to_pandas()

level2_df = load_dataset(
    "Quandela/Challenge_Swaptions",
    data_files="level-2_Missing_data_prediction/train_level2.csv",
    split="train",
).to_pandas()


def summarize_dataset(name: str, df: pd.DataFrame) -> None:
    print(f"\n{'='*80}")
    print(f"{name} — shape: {df.shape}")
    print(f"{'='*80}")

    if "Date" in df.columns:
        print("Date range:", df["Date"].min(), "->", df["Date"].max())

    numeric_df = df.select_dtypes(include=[np.number])
    print("Numeric columns:", numeric_df.shape[1])

    missing_total = int(df.isna().sum().sum())
    print("Total missing values:", missing_total)

    if numeric_df.shape[1] > 0:
        stats = numeric_df.describe().T[["mean", "std", "min", "max"]]
        print("\nStatistical summary (first 10 numeric columns):")
        print(stats.head(10))

        col_missing = df.isna().sum()
        col_missing = col_missing[col_missing > 0].sort_values(ascending=False)
        if len(col_missing) > 0:
            print("\nColumns with missing values (top 10):")
            print(col_missing.head(10))
        else:
            print("\nNo missing values detected in this dataset.")


summarize_dataset("Level-1", level1_df)
summarize_dataset("Level-2", level2_df)
```
